```python
plt.figure(figsize=(8,5))
plt.hist(df['Estimated Unemployment Rate (%)'],bins=10)
plt.title("Distribution of Unempoyment Rate")
plt.show()
```

```python
state_unemployment = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

state_unemployment.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 States by Average Unemployment Rate")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
sns.boxplot(
    x='Area',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title("Urban vs Rural Unemployment")
plt.show()
```

```python
monthly=df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(8,5))
plt.plot(monthly.index,monthly.values)

plt.title("Unemployment Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()

```

### Identify Seasonal Trendsdentify Seasonal Trends

```python
df['Month_Name'] = df['Date'].dt.month_name()
```

```python
monthly_unemployment=df.groupby("Month_Name")['Estimated Unemployment Rate (%)'].mean()

monthly_unemployment.sort_index(ascending=True)
```

```python
monthly_unemployment.plot(kind='bar', figsize=(10,5))

plt.title("Average Unemployment Rate by Month")
plt.xlabel("Month")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)

plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

numeric_df.corr()
```

```python
plt.figure(figsize=(5,3))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='Estimated Employed',
    y='Estimated Unemployment Rate (%)'
)
plt.title("Employment VS UnEmployment")
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x='Estimated Labour Participation Rate (%)',
    y='Estimated Unemployment Rate (%)'
)

plt.title("Labour Participation vs Unemployment")
plt.show()
```

```python
region_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()


region_avg.sort_values(ascending=False).head(10)
```

```python
region_avg.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Regions by Unemployment")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```
