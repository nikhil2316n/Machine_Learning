```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.groupby('Region')['Estimated Unemployment Rate (%)'].max().sort_values(ascending=False)
```

```python
df.groupby('Region')['Estimated Unemployment Rate (%)'].min().sort_values(ascending=True)
```

```python
df.groupby('Area')["Estimated Unemployment Rate (%)"].mean()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Estimated Unemployment Rate (%)'],bins=10)
plt.title("Distribution of Unempoyment Rate")
plt.show()
```

```python
state_unemployment = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

state_unemployment.sort_values(ascending=False).head(10).plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 States by Average Unemployment Rate")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
sns.boxplot(
    x='Area',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title("Urban vs Rural Unemployment")
plt.show()
```

```python
monthly=df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(8,5))
plt.plot(monthly.index,monthly.values)

plt.title("Unemployment Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()

```
